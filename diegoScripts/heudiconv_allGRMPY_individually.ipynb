```python
import sys
sys.path.append('/Users/diegodavila/miniconda3/lib/python3.7/site-packages')
import flywheel
import datetime
import pytz
from time import gmtime, strftime

fw = flywheel.Client()
proj = fw.lookup('bbl/GRMPY_822831')
sessions = proj.sessions()
sessions = [fw.get(x.id) for x in sessions]
```

```python
gear = fw.lookup('gears/fw-heudiconv')
```

```python
heur = proj.files[30]
heur
```

```python
myinputs = {
    'heuristic': heur
}
```

```python
myconfig = {
    'do_whole_project': False,
    'dry_run': False
}
```

```python
runs = {}
for ses in sessions:
    runs[ses.label] = gear.run(analysis_label="heudiconv_{}".format(strftime("%Y-%m-%d %H:%M:%S", gmtime())), destination=ses, inputs=myinputs, config=myconfig)
runs
```
