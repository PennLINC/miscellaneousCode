```python
import flywheel
import datetime
import pytz
from time import gmtime, strftime

fw = flywheel.Client()
proj = fw.lookup('bbl/GRMPY_822831')
sessions = proj.sessions()
sessions = [fw.get(x.id) for x in sessions]
```

```python
def get_latest_fmriprep(session):
    
    if session.analyses:
        
        timezone = pytz.timezone("UTC")
        init_date = datetime.datetime(2018, 1, 1)
        latest_date = timezone.localize(init_date)

        latest_run = None

        for i in session.analyses:
            gear_name = i.gear_info['name']
            state = i.job.state
            date = i.created
            if 'fmriprep' in gear_name and date > latest_date and state =='complete':
                latest_date = date
                latest_run = i
    
        return(latest_run)

    else:
        return None
```

```python
ses = sessions[0]
latest_fmriprep = get_latest_fmriprep(ses)
print(latest_fmriprep.created)
print(latest_fmriprep.gear_info)
print(latest_fmriprep.label)
```

```python
designFile = [x for x in proj.files if x.name == "task2.dsn"][0]
designFile = proj.get_file(designFile.name)
print(designFile)
```

```python
taskfile = [x for x in proj.files if x.name == "taskfile.zip"][0]
taskfile = proj.get_file(taskfile.name)
print(taskfile)
```

```python
xcp_runs = {}

for ses in sessions:
    if ses.label == "11445":
        latest_fmriprep = get_latest_fmriprep(ses)
        if latest_fmriprep is not None:
            fmriprep_out = [x for x in latest_fmriprep.files if 'fmriprep_sub' in x.name][0]

            myinputs = {
                'taskfile': taskfile,
                'designfile': designFile,
                'fmriprepdir': fmriprep_out 
            }

            myconfig = {
                'analysis type': 'task',
                'task_name': 'fracback'
            }

            xcp_runs[ses.label] = xcp.run(analysis_label="XCP_task_{}".format(strftime("%Y-%m-%d %H:%M:%S", gmtime())), destination=ses, inputs=myinputs, config=myconfig)
            #print(fmriprep_out is None, designFile is None, taskfile is None)

```
